```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split

from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Practicum/v1/df')
```

```python
df = df[['LICNSTAT','PRACTAGE','Label_LICNFELD','Label_ALEGATN1','PMT']]
```

```python
df.rename(columns = {'Label_LICNFELD':'LICNFELD', 'Label_ALEGATN1':'ALEGATN1'}, inplace = True)
```

```python
df.to_csv('/content/drive/MyDrive/Practicum/v1/prob_df', index = False)
```

```python
df['PRACTAGE'] = pd.Categorical(df.PRACTAGE)
```

```python
df = pd.get_dummies(df)
```

I. Split to train and test. We want 80-20 split, with proportional cases of 1s and 0s.

```python
df_0 = df[df['PMT'] == 0]
df_1 = df[df['PMT'] == 1]
```

```python
X_0 = df_0.drop('PMT',axis=1)
y_0 = df_0['PMT']
```

```python
# Train Test split is 0.8, 0.2
X_0_train, X_0_test, y_0_train, y_0_test = train_test_split(X_0,y_0,test_size=0.2,train_size=0.8)
```

```python
X_1 = df_1.drop('PMT',axis=1)
y_1 = df_1['PMT']
```

```python
# Train Test split is 0.8, 0.2
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X_1,y_1,test_size=0.2,train_size=0.8)
```

```python
train_0_df = pd.merge(X_0_train, y_0_train, left_index=True, right_index=True)
train_1_df = pd.merge(X_1_train, y_1_train, left_index=True, right_index=True)
```

```python
train_df = train_0_df.append(train_1_df)
train_df
```

```python
test_0_df = pd.merge(X_0_test, y_0_test, left_index=True, right_index=True)
test_1_df = pd.merge(X_1_test, y_1_test, left_index=True, right_index=True)
```

```python
test_df = test_0_df.append(test_1_df)
test_df
```

```python
#checks
train_df['PMT'].sum() + test_df['PMT'].sum()
```

```python
sev_train_df = pd.read_csv('/content/drive/MyDrive/Practicum/v1/train_df.csv')

```

```python
for i in train_df.columns:
  if i not in sev_train_df.columns and i != 'PMT':
    train_df.drop(i, axis = 1, inplace=True)
```

```python
for i in test_df.columns:
  if i not in sev_train_df.columns and i != 'PMT':
    test_df.drop(i, axis = 1, inplace=True)
```
