```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split

from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Practicum/v1/sev_df.csv')
```

Add log payment column

```python
df['LOG_TOTALPMT'] = df['TOTALPMT_ADJ'].apply(lambda x: np.log(x))
```

II. Exclude rows where outside 5 and 95 percentiles of TOTALPMT_ADJ

```python
min_pmt = df['TOTALPMT_ADJ'].quantile(0.05)
max_pmt = df['TOTALPMT_ADJ'].quantile(0.95)
```

```python
df = df[(df.TOTALPMT_ADJ > min_pmt) & (df.TOTALPMT_ADJ < max_pmt)]
```

```python
df
```

201,243 rows remain after the exclusion

III. Train test split

Identify response variables

```python
X = df.drop('TOTALPMT_ADJ',axis=1).drop('LOG_TOTALPMT', axis=1)
y = df[['TOTALPMT_ADJ','LOG_TOTALPMT']]
```

Change PRACTAGE into a categorical variable

```python
X['PRACTAGE'] = pd.Categorical(X.PRACTAGE)
```

```python
X = pd.get_dummies(X)
```

```python
X
```

Split

```python
# Train Test split is 0.8, 0.2
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,train_size=0.8)
```

```python
train_df = pd.merge(X_train, y_train, left_index=True, right_index=True)
test_df = pd.merge(X_test, y_test, left_index=True, right_index=True)
```

IV. Export to csv

Test df to be used only in final evaluation

```python
train_df.to_csv('/content/drive/MyDrive/Practicum/v1/train_df.csv', index = False)  
test_df.to_csv('/content/drive/MyDrive/Practicum/v1/test_df.csv', index = False)  
```
